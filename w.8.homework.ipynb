```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("airbnb_data.csv")
df.head()

```

```python
df.columns
```

```python
for col in df.columns:
    print("Column:", col)
    print("Null values:", df[col].isnull().sum())
    
```

```python
def numeric_summary(column):
    print(f"Col: {column.name}")
    print(f"Mean: {column.mean()}")
    print(f"Median: {column.median()}")
    print(f"Standard deviation: {column.std()}")
```

```python
for col in df.columns:
    if df[col].dtype != "object":
        numeric_summary(df[col])
```

```python
def categorical_summary(column):
    num_unique = column.nunique()
    print(f"Unique values: {num_unique}")

    if num_unique < 10:
        print(f"Value names: {column.unique()}")
```

```python
for col in df.columns:
    if df[col].dtype == "object":
        categorical_summary(df[col])
```

```python
plt.figure(figsize=(10,8))
sns.histplot(df["number of reviews"], bins=30, kde=True)
plt.title("Distribution of Number of Reviews")
plt.xlabel("Number of Reviews")
plt.ylabel("Freq")
skew = df["number of reviews"].skew()
kurt = df["number of reviews"].kurt()

plt.text(
    0.95, 0.95,
    f"skew:{skew:.3f}\nkurt:{kurt:.3f}",
    transform=plt.gca().transAxes,
    ha="right",
    va="top",
    bbox = dict(facecolor="white", edgecolor = "black")
)

plt.show
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x="room type", y="number of reviews")

plt.title("Number of Reviews by Room Type")
plt.xlabel("room type")
plt.ylabel("number of reviews")

plt.show()
```

```python
def multiplot(df, col, x_label):
    fig, (ax1, ax2)=plt.subplots(
        nrows=2,
        ncols=1,
        sharex=True,
        height_ratios=[2,8]    
    )
    sns.boxplot(data=df,x=col, ax=ax1)
    sns.histplot(data=df, x=col, ax=ax2, bins=30, kde=True)
    ax1.set(xlabel="")
    ax2.set(xlabel=x_label, ylabel="Freq")
    plt.suptitle("Distribution of number of reviews")
    plt.show()
```

```python
multiplot(df, "number of reviews", "number of reviews")
```
