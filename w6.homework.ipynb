```python
import pandas as pd

url = "https://raw.githubusercontent.com/nkmwicz/data-for-students/refs/heads/main/french-ambs.csv"
df = pd.read_csv(url)
df.head()
```

```python
df.shape
```

```python
df.loc[(df["year"] > 1588) & (df["year"] < 1592)].sum()
```

```python

df["is_permenant"] = True
temp_places = ["Geneva", "Saxony", "Tuscany", "Ferrara"]
df.loc[df["place"].isin(temp_places), "is_permenant"] = False

df.loc[df["is_permenant"] == False].shape[0]
```

```python
df.loc[df["place"] == "Geneva", "place"] = "Swiss Cantons"
df.loc[df["place"] == "Grisons", "place"] = "Swiss Cantons"

df.loc[df["place"] == "Swiss Cantons"].shape[0]
```

```python
df_agg = df.groupby("place").agg({
    "year": "count",
    "name": "nunique"
}).reset_index()

df_agg.columns = ["place", "years_present", "unique_ambs"]

df_agg = df_agg.sort_values("unique_ambs", ascending=True)
df_agg.iloc[3]["unique_ambs"]
```
